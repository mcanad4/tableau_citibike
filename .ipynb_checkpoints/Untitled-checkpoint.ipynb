```python
import os
import pandas as pd
```

```python
data1_path = os.path.join('data', 'Combined_data1.csv')
data1_df = pd.read_csv(data1_path, low_memory=False)
```

```python
# drop columns that will not be necessary for analysis
clean_data1 = data1_df.drop(columns=['tripduration','end station id','end station name','bikeid','birth year','gender'])
clean_data1
```

```python
data2_path = os.path.join('data', 'Combined_data2.csv')
data2_df = pd.read_csv(data2_path, low_memory=False)
```

```python
# drop columns that will not be necessary for analysis
clean_data2 = data2_df.drop(columns=['ride_id','rideable_type','end_station_name','end_station_id'])
clean_data2
```

```python
# rename columns for readability
clean_data2 = clean_data2.rename(columns={'started_at':'starttime',
                                          'ended_at':'stoptime', 
                                          'start_station_name':'start station name',
                                          'start_lat':'start station latitude',
                                          'start_lng':'start station longtitude',
                                          'end_lat':'end station latitude',
                                          'end_lng':'end station longitude',
                                          'member_casual':'usertype'                                        
                                         })
```

```python
clean_data2.replace({'usertype':'member'},{'usertype':'Subscriber'},inplace=True)
clean_data2.replace({'usertype':'casual'},{'usertype':'Customer'},inplace=True)
clean_data2
```

```python
# merge dataframes
frames = [clean_data1, clean_data1]

data_merge = pd.concat(frames)
```

```python
# Push the remade DataFrame to a new CSV file
data_merge.to_csv("data/data.csv", index=False, header=True)
```
